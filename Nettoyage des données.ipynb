```python
dataset=dataset.drop(columns=["quantity","iron_100g","vitamin-c_100g","ingredients_from_palm_oil_n","states_fr","cholesterol_100g","omega-3-fat_100g","serving_size","additives","ingredients_text"])
```

# Traitement des données pour chaines de charactères

```python
imputer=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
imputer=imputer.fit(dataset.iloc[:,0:3])
dataset.iloc[:,0:3]=imputer.transform(dataset.iloc[:,0:3])
```

```python
dataset.head(10)
```

# Traitement des données pour les valeurs numériques

```python
imputer=SimpleImputer(missing_values=np.nan,strategy='median')
imputer=imputer.fit(dataset.iloc[:,3:12])
dataset.iloc[:,3:12]=imputer.transform(dataset.iloc[:,3:12])
dataset
```

# Cleaning des données

```python
dataset.iloc[:,4:11]=np.where(dataset.iloc[:,4:11]>100, 100, dataset.iloc[:,4:11])
```

```python
dataset.iloc[:,3:11]=np.where(dataset.iloc[:,3:11]<0, 0, dataset.iloc[:,3:11])
```

```python
dataset.describe(include="all")
```

# Suppression des doublons dans le dataset

```python
dataset.duplicated().sum()
```

```python
dataset.drop_duplicates(keep = 'first', inplace=True)
```
